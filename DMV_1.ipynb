```python
# creating derived variables 

csv_data['full_name'] = csv_data['CONTACTFIRSTNAME'] + ' ' + csv_data['CONTACTLASTNAME'] 
```

```python
csv_data['Address'] = csv_data['ADDRESSLINE1'] + ',' + csv_data['CITY']
```

```python
csv_data['profit'] = csv_data['SALES'] - (csv_data['QUANTITYORDERED'] * csv_data['PRICEEACH'])
```

```python
# visualization
import seaborn as sns
```

```python
# countplot for sales count across the category

plt.figure(figsize=(10,8))
sns.countplot(x='PRODUCTLINE',data=csv_data)
plt.title('count of categories sales')
plt.xlabel('productline')
plt.ylabel('count')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x='PRODUCTLINE',y='SALES' ,data=csv_data,estimator='mean')
plt.title('Average sales by product category')
plt.xlabel('product line')
plt.ylabel('sales')
plt.show()
```

```python
sns.pairplot(csv_data[['PRICEEACH' , 'QUANTITYORDERED','SALES','profit']])
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='SALES',data=csv_data)
plt.title('sales distribution')
plt.ylabel('sales')
plt.show()
```

```python
plt.figure(figsize=(10,8))
merge_data.groupby('PRODUCTLINE')['SALES'].sum().plot(kind='bar',color='skyblue')
plt.title('total sales of combined data types')
plt.xlabel('prodcuts')
plt.ylabel('total sales')
plt.show()
```
